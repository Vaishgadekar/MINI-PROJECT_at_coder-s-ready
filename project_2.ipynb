### <span style ='color:red'> def function; </span>
 -with the help of data we undestand tot cases data is in the from of exponatial
 -so,we define function of exponatial for ploting graph and curve fo

```python
def exp_f(x , a, b, c):
    y = a*np.exp(b*x)+c
    return y
```

### <span style ='color:red'>curve fitting </span>

## <span style ='color:red'>“Curve Fitting" is the process of constructing a curve or mathematical function that has the best fit to a series of data points, possibly subject to constraints. </span>

```python
fit = curve_fit(exp_f, df['days'][:122],df['tot_case'][:122], p0 = [0.005,0.03,5])
fit_eq = fit[0][0]*np.exp(fit[0][1]*df['days'][:122])+ fit[0][2]
```

```python
fig = plt.figure()
ax = fig.subplots()
ax.scatter(df['days'][:122],df['tot_case'][:122], color = 'b',s =30)
ax.plot(df['days'][:122], fit_eq, color= 'r',alpha =1)
ax.set_ylabel('total cases')
ax.set_xlabel('no of days')
plt.show()
```
